## Introdução

### Diabetes
 
Segundo o Wikipedia, Diabetes mellitus, ou simplesmente diabetes, é um grupo de doenças metabólicas em que se verificam níveis elevados de glicose no sangue durante um longo intervalo de tempo. Os sintomas da elevada quantidade de glicose incluem polaquiúria (necessidade frequente de urinar) e aumento da sede (polidipsia) e da fome (polifagia). Quando não é tratada, a diabetes pode causar várias complicações. Entre as complicações agudas estão a cetoacidose, coma hiperosmolar hiperglicémico ou morte. Entre as complicações a longo prazo estão doenças cardiovasculares, acidentes vasculares cerebrais, doença renal crónica, úlceras no pé e retinopatia diabética.

A diabetes é o resultado quer da produção insuficiente de insulina pelo pâncreas, quer da resposta inadequada das células do corpo à insulina produzida.

Existem três tipos principais de diabetes:

* A diabetes mellitus tipo 1 resulta da produção de quantidade insuficiente de insulina pelo pâncreas. Este tipo era anteriormente denominado "diabetes insulino-dependente". As causas são desconhecidas.

* A diabetes mellitus tipo 2 tem origem na resistência à insulina, uma condição em que as células do corpo não respondem à insulina de forma adequada. À medida que a doença avança, pode também desenvolver-se insuficiência na produção de insulina. Este tipo era anteriormente denominado "diabetes não insulino-dependente". A principal causa é peso excessivo e falta de exercício físico.

* A diabetes gestacional é a condição em que uma mulher sem diabetes apresenta níveis elevados de glicose no sangue durante a gravidez.

### Prevenção

Tanto a prevenção como o tratamento da diabetes consistem em manter uma dieta saudável, praticar regularmente exercício físico, manter um peso normal e abster-se de fumar. Em pessoas com a doença, é importante controlar a pressão arterial e manter a higiene dos pés. A diabetes do tipo 1 deve ser tratada com injeções regulares de insulina. A diabetes do tipo 2 pode ser tratada com medicamentos por via oral como metformina e glibenclamida, com ou sem insulina. Tanto a insulina como alguns medicamentos por via oral podem causar baixos níveis de glicose no sangue. Em pessoas obesas com diabetes do tipo 2, a cirurgia de redução do estômago pode ser uma medida eficaz. A diabetes gestacional geralmente resolve-se por si própria após o nascimento do bebé. No entanto, se não for tratada durante a gravidez pode ser a causa de várias complicações para a mãe e para o bebé.

### Causas

### Sintomas

A tríade clássica dos sintomas da diabetes:

* poliúria (aumento do volume urinário);

* polidipsia (sede aumentada e aumento de ingestão de líquidos);

* polifagia (apetite aumentado).

Outros sintomas importantes incluemː

* perda de peso;

* visão turva;

* cetoacidose diabética;

* síndrome hiperosmolar hiperglicêmica não cetótica.

Pode ocorrer perda de peso. Estes sintomas podem se desenvolver bastante rapidamente no tipo 1, particularmente em crianças (semanas ou meses) ou pode ser sutil ou completamente ausente — assim como pode se desenvolver muito mais lentamente — no tipo 2. No tipo 1 pode haver também perda de peso (apesar da fome aumentada ou normal) e fadiga. Estes sintomas podem também se manifestar na diabetes tipo 2 em pacientes cuja doença é mal controlada.
Problemas de visão atingem 40% dos diabéticos insulinodependentes e e 20% dos diabéticos não insulinodependentes. Sendo mais comum em mulheres e entre os 30 aos 65 anos. Caso não seja tratado pode causar catarata, glaucoma e cegueira. Depois de 10 anos de doença, problemas de visão atingem 50% dos pacientes e depois de 30 anos atingem 90%.[29]

### Impacto
Estima-se que em 2015 cerca de 415 milhões de pessoas em todo o mundo tivessem diabetes. Cerca de 90% dos casos eram diabetes do tipo 2, o que corresponde a 8,3% da população adulta. A doença afeta em igual proporção mulheres e homens. Em 2014, a tendência sugeria que a prevalência continuaria a aumentar. 
A diabetes aumenta pelo menos duas vezes o risco de morte prematura. Entre 2012 e 2015, a diabetes foi a causa de entre 1,5 e 5 milhões de mortes anuais.
Estima-se que em 2014 o custo económico global da doença tenha sido de 612 mil milhões de dólares.

### 1.2.1 Variável Age

```python
def plot_histogram_by_target(data: pd.DataFrame, y_column: str, x_column: str, **kwargs) -> None:
    group_labels = kwargs.get('group_labels', ['Não diabetes', 'Diabetes'])
    bin_size = kwargs.get('bin_size', 5)    
    mask = data[y_column] == 0
    x_column_neg = data.loc[mask, x_column]
    x_column_pos = data.loc[~mask, x_column]
    hist_data = [x_column_neg, x_column_pos]
    fig = figure_factory.create_distplot(hist_data, group_labels, show_hist = True, bin_size = bin_size, curve_type='kde', colors=[PRIMARY_COLOR, SECONDARY_COLOR])
    fig.update_layout(margin=dict(t=60, b=0, l=0, r=0))
    fig.update_layout(title = f'Variável {x_column}')
    fig.show()
```

```python
def plot_histogram_by_target_and_sex(data: pd.DataFrame, y_column: str, x_column: str, **kwargs) -> None:
    group_labels = kwargs.get('group_labels', ['Não diabético','Diabético'])
    bin_size = kwargs.get('bin_size', 5)
    male_df = data.query('gender == "Male"')
    female_df = data.query('gender == "Female"')
    # male
    mask_male = male_df[y_column] == 0
    mask_female = female_df[y_column] == 0
    fig = make_subplots(rows=1, cols=2, subplot_titles=("Masculino", "Feminino"))
    col = 1
    showlegend = True
    for mask, df in [(mask_male, male_df), (mask_female, female_df)]:
        x_column_neg = df.loc[mask, x_column]
        x_column_pos = df.loc[~mask, x_column]
        fig.add_trace(go.Histogram(x=x_column_neg, name='Não Diabético', legendgroup='Não Diabético', showlegend=showlegend,  marker_color='#EB89B5', opacity=0.75), row=1, col=col)
        fig.add_trace(go.Histogram(x=x_column_pos, name='Diabético', legendgroup='Diabético', showlegend=showlegend, marker_color='#330C73', opacity=0.75), row=1, col=col)       
        col += 1
        showlegend = False
    fig.update_layout(margin=dict(t=60, b=0, l=0, r=0))
    fig.update_xaxes(title_text="Idade", row=1, col=1)
    fig.update_xaxes(title_text="Idade", row=1, col=2)
    fig.show()
```

```python
plot_histogram_by_target(data, 'target', 'age', bin_size=5)
```

```python
plot_histogram_by_target_and_sex(data, 'target', 'age', bin_size=5)
```

```python
def plot_pie_chart(data: pd.DataFrame, x_column: str, **kwargs) -> None:
    title = kwargs.get('title', '')
    labels = pd.value_counts(data[x_column]).index.to_list()
    bin_count = pd.value_counts(data[x_column]).to_list()
    trace = go.Pie(values=bin_count,
                   labels=labels,
                   opacity=0.8,
                   textfont=dict(size=15),
                   marker=dict(
                       colors=[PRIMARY_COLOR, SECONDARY_COLOR],
                       line=dict(color='#000000', width=1.5)))
    fig = go.Figure([trace])
    fig.update_layout(title=title, title_font_size=25, template='ggplot2')
    fig.update_layout(margin=dict(t=60, b=0, l=0, r=0))
    fig.show()    
```

```python
plot_pie_chart(data, 'gender', title='Distribuição gender')
```

```python
def plot_bar_by_target(data: pd.DataFrame, y_column: str, x_column: str, **kwargs) -> None:
    title = kwargs.get('title', '')
    col_values = list(np.unique(data[x_column]))
    neg_name, pos_name = list(np.unique(data[y_column]))
    traces = []
    target_values_percentages = {neg_name: [], pos_name: []}
    for value in col_values:
        new_data = data.query(f'{x_column} == "{value}"')
        percentages = new_data[y_column].value_counts(normalize=True).sort_index()
        percentage_values = list(percentages.values)
        target_values_percentages[neg_name].append(percentage_values[0])
        target_values_percentages[pos_name].append(percentage_values[1])
    traces.append(go.Bar(name='Diabético', x=col_values, y=target_values_percentages[pos_name], text=target_values_percentages[pos_name], textposition='auto', marker_color=SECONDARY_COLOR, width=[0.2, 0.2], opacity=0.8))
    traces.append(go.Bar(name='Não Diabético', x=col_values, y=target_values_percentages[neg_name], text=target_values_percentages[neg_name], textposition='auto', marker_color=PRIMARY_COLOR, width=[0.2, 0.2] , opacity=0.8))
    layout = go.Layout(yaxis=dict(tickformat=".2%"))
    fig = go.Figure(data=traces, layout=layout)
    fig.update_layout(barmode='stack')
    fig.update_layout(title=title, title_font_size=25)
    fig.update_layout(margin=dict(t=60, b=0, l=50, r=50))
    fig.update_traces(texttemplate="%{text:.2%}")
    fig.show()
```

```python
plot_bar_by_target(data, 'class', 'gender', title='Gênero x Diabetes (%)')
```

### Correlação
 
Como grande parte das variáveis são qualitativas, vou discretizar a variável contínua e fazer o gráfico da matriz de correlação usando uma métrica para associação entre variáveis categóricas como o ChiSquare.


```python
import scipy.stats as stats

def association_plot(data, categ_cols, num_cols):
    _data = data.copy()
    _data = data[categ_cols+num_cols]
    _data[num_cols] = _data[num_cols].apply(lambda x: pd.qcut(x, 10, labels=False))
    _data = _data.apply(lambda x: pd.factorize(x, sort=True)[0])
    correlation = _data.corr(method=lambda x, y: stats.kendalltau(x, y)[0])
    matrix_cols = correlation.columns.tolist()
    corr_array  = np.array(correlation)
    trace = go.Heatmap(z = corr_array,
                       x = matrix_cols,
                       y = matrix_cols,
                       colorscale='Viridis',
                       colorbar   = dict())
    layout = go.Layout(dict(title = 'Correlation Matrix for variables',
                            autosize = False,
                            height  = 700,
                            width   = 1200,
                            margin  = dict(r = 0 ,l = 0,
                                           t = 0,b = 0),
                            yaxis   = dict(tickfont = dict(size = 9)),
                            xaxis   = dict(tickfont = dict(size = 9)),
                           )
                      )
    fig = go.Figure(data = [trace],layout = layout)
    fig.show()
    return correlation
```

```python
CATEG_COLUMNS = ['gender', 'polyuria', 'polydipsia', 'sudden_weight_loss',
                 'weakness', 'polyphagia', 'genital_thrush', 'visual_blurring',
                 'itching', 'irritability', 'delayed_healing', 'partial_paresis',
                 'muscle_stiffness', 'alopecia', 'obesity']
NUM_COLUMNS = ['age']
TARGET = ['target']
```

```python
corr = association_plot(data.drop('class', axis=1), CATEG_COLUMNS+TARGET, NUM_COLUMNS)
```
